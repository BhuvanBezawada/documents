# Multi dimensional (MD) Workspaces

If the data is dependent on multiple coordinates, the appropriate containers in Mantid are `MDWorkspaces`. `MDWorkspaces` come in two forms [MDEventWorkspaces](http://docs.mantidproject.org/nightly/concepts/MDWorkspace.html), which are analogous to `EventWorkspaces` and the [MDHistoWorkspace](http://docs.mantidproject.org/nightly/concepts/MDHistoWorkspace.html), which is analogous to the `Workspace2D`. 

**Dimensions**

A `MDWorkspace` can have between 1 and 9 dimensions. Each dimension is defined with a name, units, and minimum/maximum extents. Click on the arrow next to the workspace name in the `Workspaces toolbox` and you can see what are the dimensions of the workspace.

**MDEvent** 

A `MDEvent` is simply a point in space defined by its coordinates, plus a signal (weight) and error.

* The `MDLeanEvent` type contains only coordinates, signal and error.
* The `MDEvent` type also contains a run index (for multiple runs summed into one workspace) and a detector ID, allowing for more information to be extracted.


**NOTE** The name is slightly misleading. An `MDEvent` does not necessarily correspond to a neutron event. It can correspond to a single event, multiple events, or even a histogram bin. 


**MDEventWorkspace storage**

An `MDEventWorkspace` can contain a large number of events, in some random order. While for an `EventWorkspace` we can sort the events according to their time of flight, doing the same for a list of `MDEvents` can pose a challenge: do we sort them by Dimension(0), then Dimension(1), Dimension(2), ..., or do we sort them by Dimension(N), Dimension(N-1), ...? Or is there a better way? This question arises when we want to select a smaller subset of the `MDEvent` list, such as a cut through along energy transfer at a given reciprocal space position.

To solve this problem, Mantid uses an adaptive gridding procedure. We start with a big `MDBox` and start storing `MDEvents`. If the number of `MDEvents` in an `MDBox` is grater than a threshold, we can split the `MDBox` into smaller parts. We can recursively repeat this procedure.

Each `MDBox` caches the total signal, error squared, and number of `MDEvents` in that box. For a given cut, if the `MDBox` is completely contained withing the cut, we can just used the cached values, and we don't need to process the individual `MDEvents`. If the `MDBox` is completely outside the cut, we don't need to process any of the `MDEvents`. The only `MDEvents` that we need to worry about are those in `MDBoxes` that intersect, but are not contained, in the cut.

**MDHistoWorkspace**

The `MDHistoWorkspace` contains data organized on a regular grid. It consists of an array of `Signal`, an array of `ErrorSquared`, and an array for number of `MDEvents` contributing to each point of the grid. 

## Create `MDEventWorkspaces`

The most common way to create an `MDEventWorkspace` is the [ConvertToMD](http://docs.mantidproject.org/nightly/algorithms/ConvertToMD-v1.html) algorithm.
ConvertToMD is an umbrella algorithm that provides a large number of conversions. The details of all the available options are outside the scope of this course.

Below is an example of the generation of a 4D MDWorkspace from a direct run on CNCS.

* Run Load with Filename CNCS_7860_event.nxs
* Run AddSampleLog with LogName=Ei, LogText=3, LogType=Number
* Run SetUB with a, b and c = 1.4165, u=[1,0,0], v=[0,1,0]
* Run SetGoniometer with Axis0=Psi,0,1,0,1
* Now Run ConvertToMD with QDimensions=Q3D, QConversionScales=HKL, dEAnalysisMode=Direct, MinValues=[-3,-3,-3,-1], MaxValues=[3,3,3,3]

or run the following in either the script interpreter or the script window:

    ws = Load(Filename='CNCS_7860_event')
    AddSampleLog(Workspace=ws,LogName='Ei',LogText='3.0',LogType='Number')
    SetUB(Workspace=ws,a='1.4165',b='1.4165',c='1.4165',u=[1,0,0],v=[0,1,0])
    AddSampleLog(Workspace=ws,LogName='Psi',LogText='0.0', LogType='Number')
    x = ws.getRun().getLogData('Psi')
    SetGoniometer(Workspace=ws,Axis0='Psi,0,1,0,1')
    mdws = ConvertToMD(InputWorkspace=ws,   
                       QDimensions='Q3D',
                       QConversionScales='HKL',
                       dEAnalysisMode='Direct',
                       MinValues=[-3,-3,-3,-1],
                       MaxValues=[3,3,3,3])
    plotSlice(mdws, xydim=["[H,0,0]","[0,K,0]"], 
              slicepoint=[0,0], colorscalelog=True )

## Visualizing MDWorkspaces - SliceViewer

```python
Image(filename='images/SliceViewer.png')
```

To produce the dataset for the figure above, run the `CreateSXDNaClQLab.py` script in the training dataset

The `SliceViewer` provides 2D slices through `MDWorkspaces`. It is designed to provide very fast, interactive projections on large datasets. The `SliceViewer` is also furnished with a number of 2D tools to aid further visualisation and analysis.

Features include:

* LineViewer for creating arbitrary 1D integrated cuts as separate workspaces
* PeaksViewer for over-plotting one or more PeaksWorkspaces in the Q coordinate frame
* Control via python SliceViewer Python Interface
* In-situ rebinning according to slice location
* Zooming and panning control

To start the `SliceViewer`, right click on a `MDWorkspace` in the `Workspaces toolbox` and select `Show Slice Viewer`

**Controls:**

** 1. Dimension Selection**

Allows you to choose which dimensions in the workspace are to be plotted X and Y on the plot. All other dimensions are provided as sliders. Each dimension has an X button and a Y button. Selecting one of these will make this dimension either the X or Y of the plot.

** 2. Slice Point Selection **

Sliders allow you to choose the slice location. The plot will immediately update to show a projection at the new slice location. N-2 dimensions are represented with sliders. 

** 3. Line Integration Mode**

Selecting this button will allow you to draw a line on the current slice. This will also open the `LineViewer` panel. The line can be controlled in the following ways:

* Start and end positions
* Width in plane
* Thickness (into plane)
* Number of bins

The controls may be updated either graphically, for example by grabbing the start position point and moving it with the left-click mouse option, or by updating the coordinates in the LineViewer panel.

Hitting the `Apply` button results in the generation of a new Workspace which is one-dimensional. 

```python
Image(filename='images/LineViewer.png')
```

**4. Dynamic rebinning**

The `SliceViewer` shows the box structure of `MDEventWorkspaces`, with boxes colored according to the signal in the box. The box structure might be irregular. In order to see the signal on a regular grid, use the `dynamic rebinning` buttons.


**5. Peaks Viewer**

The `PeaksViewer` is a tool for Single Crystal Diffraction. It allows you to overlay one or more `PeaksWorkspaces` over the `MDWorkspace` to which it belongs. `PeaksWorkspaces` are typically generated by running algorithms, such as `FindPeaksMD`.

* Slice through the `MDWorkspace` and the `PeaksWorkspaces` simultaneously in the `SliceViewer` plot
* Select peaks from the peaks list in the `PeaksViewer`, and cause the `SliceViewer` to zoom into that exact location on the plot
* Show integration regions and optionally integration background regions
* `PeaksViewer` auto selects the coordinates frame for the `PeaksWorkspaces` based on the `MDWorkspace`
* Sort the `PeaksWorkspaces` in-situ, for example by greatest intensity

Amongst other usages, this tool therefore allows you to inspect the quality of the peak finding, or peak integration operations.

## Normalization choices

Both `SliceViewer` and `VSI` allow users to choose a normalization option.

* **No Normalization** - choose this option if your data is on a regular grid (`MDHistoWorkspace`) and it is already normalized (`MDNorm...` algorithms)
* **Volume Normalization** - choose this option when using `MDEventWorkspaces` when looking at raw neutron counts. The number of `MDEvents` is about the same in every box (they will have about the same color), but the boxes around Bragg peaks are much smaller. For this normalization, the color is proportional to the density of counts.
* **Number of Events Normalization** - Use this option only for inelastic scattering, if the workspace that you converted to MD is a `Workspace2D`. Each of the `Workspace2D` bins is a `MDEvent`, so choosing this normalization means that the color is proportional to the average intensity in the `MDBox`, not the summed intensity. This is similar to what MSlice and Horace programs are doing.
