# Custom Interfaces

Some analysis workflows are too complex, need more user input or need a better way to present the results than a single algorithm interface can provide. This is where custom interfaces come in. These are pluggable interfaces for Mantidplot that are loaded at run time to give a much richer interface to work with for your data analysis. Several interfaces have now been developed to handle different aspects of the reduction and analysis workflow for data from the various scientific techniques.
