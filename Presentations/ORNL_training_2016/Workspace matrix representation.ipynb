# Workspace matrix representation

Once a matrix workspace is loaded into MantidPlot, one can use tha matrix view to look at histogram or event data. To activate this view, there are several options:
* Double click the workspace name
* Click and drag the workspace to the main window area
* Right click and select `Show Data`

Each row contains value from a single spectra. 

You can select tabs to look at `X` values, `Y` values (default), or `Errors`

ISIS instruments allow loading of the monitors in the same workspace as the data from the detectors. Monitor data appears on a yellow background, detector data has a cyan background.

**Histogram vs point data**
* The raw data that we load at SNS and ISIS is in a histogram mode, meaning that the `X` coordinate in the workspace stores bin boundaries. That means that there is one more `X` value that there are `Y` or `Errors` values. The columns for `Y` and `Errors` are labeled by the bin centers.
* Workspaces containing point data have the same number of `X`, `Y`, and `Errors` values. Outputs from fitting have this form


**Detector table**
* Some instruments refer to particular spectra by a unique number, the *`spectra number`*. 
* Each spectra can be the result of measurements from one or more detectors.
* To get information about spectra and detectors, one can right click on a workspace and select `Show Detectors`
