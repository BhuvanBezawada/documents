# Matrix workspaces

Workspaces come in several forms, but the most common by far is the [MatrixWorkspace](http://docs.mantidproject.org/nightly/concepts/MatrixWorkspace.html) which represents XYE data for one or more spectra. The MatrixWorkspace itself can be sub-grouped into [EventWorkspaces](http://docs.mantidproject.org/nightly/concepts/EventWorkspace.html) and [Workspace2Ds](http://docs.mantidproject.org/nightly/concepts/Workspace2D.html). 

## Workspace2D

A Workspace2D consists of a workspace with 1 or more spectra. Typically, each spectrum will be a histogram. For each spectrum X, Y (counts) and E (error) data is stored as a separate array.

Each workspace has two axes, the Spectrum axis and the X-axis. Where an axis holds a known unit type, it may be converted to another set of units.

**Ragged Workspaces**

Converting x-axis units can lead to a ragged workspace, in which bin boundaries are not consistent across the spectra. Some algorithms will not accept input workspaces that are ragged. The fix to this is to apply Rebin to the Ragged workspace structure, as the example below shows.
* run `Load` on *GEM38370_Focussed.nxs* setting the `OutputWorkspace` to be *ws*
* run `ConvertUnits` on *ws* setting `OutputWorkspace` to *lambda*, `Target` to *Wavelength*, `EMode` to *Elastic*. Plotting this in the `Color Fill Plot` demonstrates the ragged X-bins.

```python
Image(filename='images/Ragged.png')
```

* run `Rebin` on *lambda* setting `Params` to 0.5 and `OutputWorkspace` to *Rebinned*. Plotting this in the `Color Fill Plot` demonstrates that uniform binning across all spectra has been achieved.

## Event Workspaces

An EventWorkspace stores information about each individual event observation in detectors. More specifically, at a neutron spallation source, this means that the time of arrival and detector ID of each individual neutron is recorded. Only fairly recent advances in computer and acquisition hardware have made storing this detailed knowledge a practical solution. For example at the SNS facility all detector data, and some collected in monitors, are stored in this way.

**Nomenclature:**

**Event** specifies “when” and “where”

**Event list** is a list of events belonging to the same spectra. They do not need to be sorted. Adding two spectra means appending their event lists.

**Weighted event** is a special type of event, that contains a weighting factor. For example, when we subtract spectra *s2* from spectra *s1*, we apply a wight of -1 to all events in *s2* before appending to *s1*. Multiplying a simple event with a constant means that we attach that constant as a weight to that event.

**Pulse time** – when the proton pulse happened in absolute time

**Time-of-flight** – time for the neutron to travel from moderator to the detector 

All event workspaces have a histogram representation, that is calculated on the fly (when needed). When we load raw event nexus files at SNS, the default histogram representation contains only one bin per spectrum.

## Rebinning

Rebinning a `Workspace2D` is a one-way process when the rebinning leads to a coarser structure. 

Rebinning an `EventWorkspace` is essentially free and can be conducted in-place. This is because the data does not need to change, only the overlaying histogramming.
