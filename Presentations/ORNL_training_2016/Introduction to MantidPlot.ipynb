# Introduction to MantidPlot

## What is MantidPlot?

* The Mantid project is a large collaboration between ISIS, ORNL, ESS, with contributions from other facillities to analyze neutron and muon experiments
* Multi platform (Linux, Windows, MacOS)
* Written in C++ and Python

```python
from IPython.display import Image
Image(filename='images/MantidFramework.png') 
```

* Workspaces contain data, informations about sample environment, instrument geometry
* Algorithms create, modify, delete workspaces
* Algorithms can be provided by users
* Interacting with the Framework occurs through the Python API or through MantidPlot
* Custom interfaces are available in MantidPlot

## How to launch MantidPlot?

* on the analysis.sns.gov machines open a terminal and type **`MantidPlot`**
* alternatively you can type **`mantidplot`**
* **NDAV policy** - MantidPlot points to the latest stable version of MantidPlot, at the beginning of the neutron cycle

## Check the version of MantidPlot

**`mantidplot --version` **

3.5.2 (Wed, 30 Mar 2016)

* You can choose a newer stable version **`mantidplot36`**, or the latest development version **`mantidplotnightly`**

## Python with Mantid

* Can be run within MantidPlot
* Can be run from within **`python`**, **`ipython`**, **`ipython notebook`**
* For better performance, use **`mantidpython`** or **`mantidpython notebook`**
