# Workspace content

All workspaces contain some form of data. We've seen the type of data contained in matrix workspaces. There are other workspace types as well.
* **Group workspaces** contain multiple workspaces. We've seen such example in multiple file loading. Either the whole group or individual members can be processed using algorithms. 
* **Table workspaces** stores data as cells. Columns determine the type of the data, for example double precision float, while each entry appears as a new row. This is analogous to a Microsoft Excel Spreadsheet. Fit parameters are stored in `table workspaces`. To see the content of a `table workspace` just double click it in the `workspaces panel`, or right click and choose `Show Data`.
* **PeaksWorkspace** is a special type of `TableWorkspace` with additional support for Single Crystal peaks. 
* **MDWorkspaces** contain data that can have more than one coordinate

## Sample Logs

Many workspaces contain information about sample environment logs. To see these logs, right click on a workspace in the `workspace panel`, and select `Sample Logs...`. A new window will open, where sample logs are listed.

Most of the logs have some time dependence, so they are stored as `number series`. If you do a single click on one of the logs, the `Log Statistics` part of the window will show some properties of the selected log.

If one double clicks on a log, or click the `Import selected log` button, a table will show up containg the log, and for `number series` they will be plotted as well.

## History

Mantid keeps the entire history of all algorithms applied to workspaces. Not only does this allow you to audit the data reduction and analysis, it also provides the means to extract a re-executable python script from the GUI.

* Right click on HYS_11388_event and select 'Show History'. This will open up the Algorihm History window. You should see something like:  

```python
Image(filename='images/AlgorithmHistory.png')
```

This reveals the history done to this workspace, i.e. Load to load the workspace, SumSpectra, and a few Rebins.

To replay the history do:
* From the main MantidPlot menu select `View->Script Window`, this opens the `MantidPlot: Python Window`
* Go back to `Algorithm History` window and press the `Script to Clipboard` button and close the Algorithm History window
* Flip to the script window and paste what was copied to the clipboard into the script window
* Close the HYS_11388_event-1 plot window
* Delete the HYS_11388_event workspace from the `Algorithms panel`
* To recreate the work you have just done script window execute the script via `Execute->Execute All` on the script window.
* To verify that history has been replayed either look at history or plot spectrum again 

Alternatively, you could have pasted the script in the `Script Interpreter` window and press `Enter`

**Note:** `Save...` algorithms will not show up in history
