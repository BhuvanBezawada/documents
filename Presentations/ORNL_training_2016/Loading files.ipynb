# Loading Files

* There are many file types that Mantid can read:
    - event nexus files
    - processed nexus files
    - nxspe
    - gsas
    - ASCII data
    - live streams
    - ...

* each of these file format has a loader algorithm, such as **`LoadEventNexus`**
* for most common formats, the **`Load`** algorithm can figure out wich loader to use
* type `Load` in the algorithm box, or use the **`Load`** button in the workspaces panel
* One can use the File->Load->File menu

## Filename hints

* It is not always necessary to input the full path if you try to load the raw event file
* if you have a default instrument, you can just input the run number
* for other instruments you can just type the short instrument name and the run number (`HYS12345` or `HYS_12345`)

**Note** the file must be in your search path

## Archive search

* if archive search is turned on (`File->Manage User Directories`) Mantid will search the `/SNS/instrumnet/` directories for the requested file

## Loading multiple files

* one can load multiple files to be processed either together or separately
* `HYS12345,12346` will load runs 12345 and 12346 in separate workspaces
* `HYS12345+12346` will load **and** add together the two runs
* for loading separately a range of run you can use something like `HYS12345:12349:2` to load runs 12345,12347,12349
* if you want to add together the same files, use `"-"` (`HYS12345-12349:2`)
