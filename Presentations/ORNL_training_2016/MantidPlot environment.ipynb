**Workspace toolbox**
* workspaces are data containers in Mantid
* the workspace panel contains the list of workspaces in memory
* some workspaces might be hidden
* workspaces have a context sensitive menu
* clicking on the arrow (or `+` on some platforms) show more information about the workspace (type, title, size, lattice parameters, ...)

**Algorithms toolbox**
* type alorithm name to run
* select algorithm from list

** Results Log panel**
* display execution status for algorithms (such as duration, errors, etc)
* One can select a `Log Level` (from `debug` to `error`)

**Fit toolbox**
* input fitting functions and parameters

** IPython interpreter and Script window**
* allow interaction with `MantidPlot` using `python`
* will be discussed later in more detail

** Preferences**
* in View->Preferences
* allows changing `Mantid` and `MantidPlot` configuration (default instrument, custom algorithms locations, plotting preferences)
